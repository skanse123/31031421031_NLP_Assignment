```python
from bs4 import BeautifulSoup
data_file = "/content/sample_data/uk_england_london_130_queensgate_london_apartments"
csv_file = "/content/sample_data/uk_england_london_130_queensgate_london_apartments.csv"
with open(data_file) as txt_file:
    data = txt_file.read()
soup = BeautifulSoup(data, 'lxml')
csv_data = []
csv_data.append(["date","review","short_review"])
for doc_tag in soup.find_all("doc"):
    raw_data = []
    raw_data.append(doc_tag.find("date").text)
    raw_data.append(doc_tag.find("review").text)
    raw_data.append(doc_tag.find("short_review").text)
    csv_data.append(raw_data)
import csv

def write_csv(file,data):
    with open(file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)
write_csv(csv_file,csv_data)
```

```python
import pandas as pd
df = pd.read_csv(csv_file)
df.head(5)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
threshold_lower = 0.4
threshold_upper = 0.85
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

scores = []
rating = []

for i in df["review"]:
    if sid.polarity_scores(i)["compound"] < threshold_lower:
        rating.append("Negative")
    elif sid.polarity_scores(i)["compound"] < threshold_upper:
        rating.append("Neutral")
    else:
        rating.append("Positive")
    scores.append(sid.polarity_scores(i)["compound"])
df["Comment_Rating"] = scores
df["Feedback"] = rating
df.head(10)
```
